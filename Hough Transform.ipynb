```python
import sys
import cv2 
import numpy as np
import glob
import matplotlib.image as mpimg
```

```python
x_files =  glob.glob('..\\set_images\\*.png')
x_file = x_files[9]
```

```python
x_file
```

```python
for i,x_file in enumerate(x_files[:3]):
    img = cv2.imread(x_file,0)
    img = cv2.medianBlur(img,7)
    cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)

    circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1,400,
                                param1=100,param2=40,minRadius=30,maxRadius=3000)


    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        # draw the outer circle
        cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
        # draw the center of the circle
        cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),3)
        

#     fig = plt.figure(figsize=(60,8))
#     ax = fig.add_subplot(4,4,i)
#     ax.set_title(x_file)
    
#     ax.imshow(cimg)
#     fig.tight_layout()

    cv2.imshow('detected circles',cimg)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
